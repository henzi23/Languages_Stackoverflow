To help in understanding this we can look at the column YearsCodedJob.

```python
# Get value counts for YearsCodedJob
df_prof_emp.YearsCodedJob.value_counts()
```

The column has quite a few categories.  For analysis I'd liek to decrease this to only a few:
* New : 0-3 years
* Mid : 3-10 years
* Experienced : >10 years

Thus we is need to recode the column.

```python
# Recode YearsCodedJob

# Create a dictionary for the recoding
recode_dict = {'1 to 2 years':'New', '2 to 3 years':'New', 'Less than a year':'New',
               '3 to 4 years':'Mid','4 to 5 years':'Mid','5 to 6 years':'Mid',
               '6 to 7 years':'Mid','7 to 8 years':'Mid','8 to 9 years':'Mid',
               '9 to 10 years':'Mid','10 to 11 years':'Experienced','11 to 12 years':'Experienced',
               '12 to 13 years':'Experienced','13 to 14 years':'Experienced',
               '14 to 15 years':'Experienced','15 to 16 years':'Experienced',
               '16 to 17 years':'Experienced','17 to 18 years':'Experienced',
               '18 to 19 years':'Experienced','19 to 20 years':'Experienced',
               '20 or more years':'Experienced'}

# Recode
df_prof_emp['YearsCodedJob'].replace(recode_dict,inplace=True)

# Check result
df_prof_emp.YearsCodedJob.value_counts()
```

The recoding looks successful.  Now that we have our updated dataset, Let's move into analysis.

## Analysis

The analysis will center around answering 3 questions:
* Question 1: What programming languages do professional developers most commonly use?
* Question 2: What programming languages do professional developers want to use in the next year?
* Question 3: Does the amount of experience (years in coding job) affect the distribution for commonly used and want to use programming language?


### Question 1: What programming languages do professional developers most commonly use?
We will analyze the HaveWorkedLanguage column to gain understanding about what languages have been utilized by employed, professional developers.  We will start by breaking down the semicolon seperated entries into lists.  Then add up number of rows are calling out each language.

```python
# Function to get counts of entries by programming language

def language_counts(df,col):
    '''
    Function to get counts of entries that answered each programming langauage.
    
    Inputs:
        df- pandas dataframe with survey data
        col- string with name of column to be investigated
    
    Outputs:
        pandas dataframe with columns for language, count of entries, and % of entries
            sorted by counts.
    '''
    
    # split string of entries into list of langauges
    df[col] = df[col].str.replace(' ','')
    val_count = df[col].value_counts().reset_index()
    val_count['index'] = val_count['index'].str.split(';')
    
    # create list of unique language entries
    entries = val_count['index'].tolist()
    entries = [item for items in entries for item in items]  #un-nest list of lists
    entries = list(set(entries))
    
    out_df = defaultdict(int)
    
    #loop through Languages
    for language in entries:
        for idx in range(val_count.shape[0]):
            if language in val_count['index'][idx]:
                out_df[language] += int(val_count[col][idx])
    out_df = pd.DataFrame(pd.Series(out_df)).reset_index()
    out_df.columns = ["Language","Count"]
    out_df['Percent'] = out_df.Count/df[col].count() * 100
    out_df.sort_values("Count",ascending=False, inplace=True)
    return out_df
    
```

```python
# Get language counts for HaveWorkedLanguage
have_worked = language_counts(df_prof_emp,col='HaveWorkedLanguage')
have_worked
```

The breakdown of counts by programming language is shown above.  Unsurprisingly we see that Java, SQL, C, and Python near the top.  Let's plot the data to be a bit more visually appealing and for ease of comparisons.

```python
# plot the top 20
plt.rcParams["figure.figsize"]=10,5

plt.bar(have_worked.Language[0:20],have_worked.Percent[0:20])
plt.xticks(rotation=65)
plt.ylabel("% of respondents")
plt.title("Top 20 Languages Used by Developers",size=20);

plt.savefig('use.jpg', bbox_inches='tight')
```

### Question 2: What programming languages do professional developers want to use in the next year?

We will now analyze the WantWorkLanguage in a similar fashion as HaveWorkedLanguage to gain an understanding of the languages developers want to use in the next year.

```python
# Get language counts for WantWorkLanguage
want_work = language_counts(df_prof_emp,col='WantWorkLanguage')
want_work
```

Again we see Java, SQL, C, and Python near the top, but now we see some other languages like TypeScript and Go moving up the list.  As before we will plot the data.

```python
# PLot Top 20
plt.rcParams["figure.figsize"]=10,5

plt.bar(want_work.Language[0:20],want_work.Percent[0:20])
plt.xticks(rotation=65)
plt.ylabel("% of respondents")
plt.title("Top 20 Developers Want to Use",size=20);

plt.savefig('want.jpg', bbox_inches='tight')
```

### Question 3: Does the amount of experience (years in coding job) affect the distribution for commonly used and want to use programming language?

To answer this question we will combine analysis like that done for question 1 and 2 broken down by our recoded column YearsCodedJob.

```python
# Analyze HaveWorkedLanguage by experience

# Create function to generate language counts by experience

def language_counts_experience(df,col,experience):
    '''
    Function to get counts of entries that answered each programming langauage by 
    groupings from YearsCodedJob.
    
    Inputs:
        df- pandas dataframe with survey data
        col- string with name of column to be investigated
        experience - grouping from column YearsCodedJob to analyze
    
    Outputs:
        pandas dataframe with columns for language, count of entries, and % of entries
            sorted ascending by percent.
    '''
    exp_df = df[df.YearsCodedJob == experience]
    output = language_counts(exp_df,col)
    output.sort_values("Percent",ascending=True, inplace=True)
    return output
```

```python
# Analyze HaveWorkedLanguage by experience

new_have_worked = language_counts_experience(df_prof_emp,'HaveWorkedLanguage','New')
mid_have_worked = language_counts_experience(df_prof_emp,'HaveWorkedLanguage','Mid')
experienced_have_worked = language_counts_experience(df_prof_emp,'HaveWorkedLanguage','Experienced')
```

To help analyze we will plot these three next to each other.

```python
# Plot HaveWorkedLanguage by experience

fig, ax = plt.subplots(ncols=3, figsize=(15,5))

fig.suptitle("Programming Languages Used by Developers by Coding Experience",size=20)
fig.subplots_adjust(top=0.85)
fig.subplots_adjust(wspace=0.5)

new_have_worked[-10:].plot.barh(x='Language',y='Percent',color='c',legend=False, ax=ax[0])
ax[0].set_title('0-3 yrs in coding job')
ax[0].set_xlabel('% of respondents')
ax[0].set_ylabel('')

mid_have_worked[-10:].plot.barh(x='Language',y='Percent',color='c',legend=False, ax=ax[1])
ax[1].set_title('0-10 yrs in coding job')
ax[1].set_xlabel('% of respondents')
ax[1].set_ylabel('')

experienced_have_worked[-10:].plot.barh(x='Language',y='Percent',color='c',legend=False, ax=ax[2])
ax[2].set_title('10 or more years in coding job')
ax[2].set_xlabel('% of respondents')
ax[2].set_ylabel('');

plt.savefig('used_by_experience.jpg', bbox_inches='tight')
```

```python
# Analyze WantWorkLanguage by experience

new_want_work = language_counts_experience(df_prof_emp,'WantWorkLanguage','New')
mid_want_work  = language_counts_experience(df_prof_emp,'WantWorkLanguage','Mid')
experienced_want_work  = language_counts_experience(df_prof_emp,'WantWorkLanguage','Experienced')
```

Let's plot the results to get a visual comparison.

```python
# Plot WantWorkLanguage by experience

fig, ax = plt.subplots(ncols=3, figsize=(15,5))

fig.suptitle("Programming Languages Want to be Used by Coding Experience",size=20)
fig.subplots_adjust(top=0.85)
fig.subplots_adjust(wspace=0.5)

new_want_work[-10:].plot.barh(x='Language',y='Percent',color='c',legend=False, ax=ax[0])
ax[0].set_title('0-3 yrs in coding job')
ax[0].set_xlabel('% of respondents')
ax[0].set_ylabel('')

mid_want_work[-10:].plot.barh(x='Language',y='Percent',color='c',legend=False, ax=ax[1])
ax[1].set_title('0-10 yrs in coding job')
ax[1].set_xlabel('% of respondents')
ax[1].set_ylabel('')

experienced_want_work[-10:].plot.barh(x='Language',y='Percent',color='c',legend=False, ax=ax[2])
ax[2].set_title('10 or more years in coding job')
ax[2].set_xlabel('% of respondents')
ax[2].set_ylabel('');

plt.savefig('want_by_experience.jpg', bbox_inches='tight')
```

Again we do not see any major shifts.  We do see some minor trends like Python shows higher for developers with less than 10 years of coding job expereince.  Additionally less propensity towards C# with less experience.  For all groupings we see some newer languages like Go and Swift showing up.
